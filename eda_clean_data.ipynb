```python
data = pd.read_csv('data/labelled_data.csv')
data['timestamp'] = pd.to_datetime(data['timestamp'])
```

```python
# Print the first few rows of the merged data
print(display(data.head()))

# Print information about the merged dat
print(data.info())
```

## FILL MISSING VALUES

```python
print(data.isnull().sum())
```

```python
## MAKE DFS FOR MISSING VALUES
missing_values = data[data.isnull().any(axis=1)]
print(display(missing_values.head()))
```

### SLEEP SCORE

```python
sleep_score_missing = missing_values[missing_values['sleep_score'].isnull()]
print(display(sleep_score_missing.head(20)))

```

```python
# FIND AVERAGE SLEEP SCORE FOR MONTH 2, 3 and 4
avg_sleep_score_2 = data[data['timestamp'].dt.month == 2]['sleep_score'].median()
avg_sleep_score_3 = data[data['timestamp'].dt.month == 3]['sleep_score'].median()
avg_sleep_score_4 = data[data['timestamp'].dt.month == 4]['sleep_score'].median()

print(avg_sleep_score_2, avg_sleep_score_3, avg_sleep_score_4)

#print average sleep score general
print(data['sleep_score'].median())
print(data['sleep_score'].mean())
```

```python
# Fill missing sleep score with median sleep score
data['sleep_score'] = data['sleep_score'].fillna(data['sleep_score'].median())
```

```python
print(data['sleep_score'].isnull().sum())
```

## HRV AVERAGE

```python
#print average hrv general
print(data['hrv_avg'].median())
print(data['hrv_avg'].mean())

```

```python
# print correlation between hrv_avg and sleep_score
print(data['hrv_avg'].corr(data['sleep_score']))
```

```python
#filter tiers of sleep scores, and their mean hrv_avg
data['sleep_score_tier'] = pd.cut(data['sleep_score'], bins=[0, 50, 70, 90, 100], labels=['Poor', 'Fair', 'Good', 'Excellent'])
print(data['sleep_score_tier'].value_counts())
print(data.groupby('sleep_score_tier')['hrv_avg'].mean())
```

```python
# Fill misisng hrv_avg with mean hrv_avg for each sleep score tier
data['hrv_avg'] = data['hrv_avg'].fillna(data.groupby('sleep_score_tier')['hrv_avg'].transform('mean'))
```

```python
print(data['hrv_avg'].isnull().sum())
```

```python
data.head()
```

## SPO2

```python
# print spo2 missing values df
print(display(data[data['spo2'].isnull()]))
```

```python
# print average spo2
print(data['spo2'].median())
print(data['spo2'].mean())

#min and max spo2
print(data['spo2'].min())
print(data['spo2'].max())
```

```python
#fill using median
data['spo2'] = data['spo2'].fillna(data['spo2'].median())
```

## BODY BATTERY

```python
print(display(data[data['body_battery'].isnull()]))
```

```python
# Create time_of_day column
data['time_of_day'] = pd.cut(data['timestamp'].dt.hour, 
                            bins=[0, 6, 12, 18, 22, 24], 
                            labels=['Night', 'Morning', 'Afternoon', 'Evening', 'Night'],
                            include_lowest=True,
                            ordered=False)  # Add this parameter

# Print the distribution of time periods
print(data['time_of_day'].value_counts())
```

```python
# Calculate average body battery by time of day
avg_body_battery = data.groupby('time_of_day')['body_battery'].mean().round(2)

# Print the results
print("\nAverage Body Battery by Time of Day:")
print(avg_body_battery)
```

```python
# Fill missing body battery with average body battery by time of day
data['body_battery'] = data['body_battery'].fillna(data.groupby('time_of_day')['body_battery'].transform('mean'))
print(data['body_battery'].isnull().sum())
```

## HRV STRESS AND RESP

```python
# Drop missing values
data = data.dropna()
```
